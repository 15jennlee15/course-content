```python
#@title Helper Functions
def plot_density_image(x, y, theta, sigma=1, ax=None):
  # plot the probability density of p(y|x,theta)
  if ax is None:
    fig, ax = plt.subplots()

  xmin, xmax = np.floor(np.min(x)), np.ceil(np.max(x))
  ymin, ymax = np.floor(np.min(y)), np.ceil(np.max(y))
  xx = np.linspace(xmin, xmax, 50)
  yy = np.linspace(ymin, ymax, 50)

  surface = np.zeros((len(yy), len(xx)))
  for i, x_i in enumerate(xx):
    surface[:, i] = stats.norm(theta*x_i, sigma).pdf(yy)

  ax.imshow(surface, origin='lower', aspect='auto', vmin=0, vmax=None, 
            cmap=plt.get_cmap('Wistia'), 
            extent=[xmin, xmax, ymin, ymax])
  ax.set(xlabel='x', ylabel='y')


def solve_normal_eqn(x, y):
  """Solve the normal equations to produce the value of theta_hat that minimizes
    MSE.  
    
    Args:
    x (ndarray): An array of shape (samples,) that contains the input values.
    y (ndarray): An array of shape (samples,) that contains the corresponding
      measurement values to the inputs.
    thata_hat (float): An estimate of the slope parameter. 

  Returns:
    float: The mean squared error of the data with the estimated parameter.
  """
  theta_hat = (x.T @ y) / (x.T @ x)
  return theta_hat
```

In the previous tutorial we took a first approximation assumption, that the data was drawn from a linear relationship with noise added, and found an effective approach for estimating model paramters based on minimizing the mean squared error.

In that case we treated the noise as simply a nuisance, but what if we factored it directly into our model?

Recall our linear model:

\begin{align}
y = \theta_0 + \theta_1 x + \eta.
\end{align}

The noise component $\eta$ is often modeled as a random variable drawn from a Gaussian distribution (also called the normal distribution).

The Gaussian distribution is described by its [probability density function](https://en.wikipedia.org/wiki/Probability_density_function) (pdf)
\begin{align}
\mathcal{N}(x; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{1}{2\sigma^2}(x-\mu)^2}
\end{align}

and is dependent on two parameters: the mean $\mu$ and the variance $\sigma^2$. We often consider the noise signal to be "white noise" which is to say that it is Gaussian with zero mean and unit variance:

\begin{align}
\eta \sim \mathcal{N}(0, 1).
\end{align}

Use the explorer widget below to see how varying the $\mu$ and $\sigma$ parameters change the location and shape of the samples.


```python
#@title Gaussian Distribution Explorer
@widgets.interact(mu=widgets.FloatSlider(0.0, min=-2.0, max=2.0),
                  sigma=widgets.FloatSlider(1.0, min=0.5, max=2.0))
def plot_normal_dist(mu=0, sigma=1):
  rv = stats.norm(mu, sigma)
  x = np.linspace(-5, 5, 100)
  y = rv.pdf(x)
  samples = rv.rvs(1000)

  fig, ax = plt.subplots()
  ax.hist(samples, 20, density=True, histtype='stepfilled', alpha=0.8,
          label='samples')
  ax.plot(x, y, color='limegreen', linewidth=2, label='pdf')
  ax.vlines(mu, 0, rv.pdf(mu), colors='orange', label='$\mu$')
  ax.vlines([mu-sigma, mu+sigma], 0, rv.pdf([mu-sigma, mu+sigma]), colors='red',
            label='$\sigma$')
  ax.set(xlabel='x', ylabel='probability density', xlim=[-5, 5], ylim=[0, 1.0])
  ax.legend()
```


### Probabilistic Models

Now that we have a model of our noise component $\eta$ as random variable, how do we incorporate this back into our original linear model from before? Consider again our simplified model $y = \theta x + \eta$ where the noise has zero mean and unit variance $\eta \sim \mathcal{N}(0, 1)$. We can now also treat $y$ as a random variable drawn from a Gaussian distribution where $\mu = \theta x$ and $\sigma^2 = 1$:

\begin{align}
y \sim \mathcal{N}(\theta x, 1)
\end{align}

which is to say that the probability of observing $y$ given $x$ and parameter $\theta$ is
\begin{align}
p(y|x,\theta) = \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}(y-\theta x)^2}.
\end{align}

### Likelihood Estimation

Now that we have our probabilistic model, we turn back to our original challenge of finding a good estimate for $\theta$ that fits our data. Given the inherent uncertainty when dealing in probabilities, we talk about the [likelihood](https://en.wikipedia.org/wiki/Likelihood_function) that some estimate $\hat \theta$ fits our data. The likelihood function $\mathcal{L(\theta)}$ is equal to the probability density function parameterized by that $\theta$:

\begin{align}
\mathcal{L}(\theta|x,y) = p(y|x,\theta) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{1}{2\sigma^2}(y-\theta x)^2}.
\end{align}

Now that we have our likelihood function, we can produce probabilities that our estimate $\hat{\theta}$ generated the provided observations. Let's try with one of the samples from our dataset.

```python
likelihood(1.0, x[1], y[1])
```

We should see that $\mathcal{L}(\theta=1.0|x=2.1,y=3.7) \approx 0.11$. So far so good, but how does this tell us how this estimate is better than any others?

When dealing with a set of data points, as we are with our dataset, we are concerned with their joint probability -- the likelihood that all data points are explained by our parameterization. Since we assume the noise affects each output independently, we can represent this as

\begin{align}
\mathcal{L}(\theta|x,y) = \prod_{i=1}^N \mathcal{L}(\theta|x_i,y_i).
\end{align}

Such a product is numerically unstable (multiplying small values together leads to underflow), so taking the logarithm of the likelihood helps by transforming products into sums:

\begin{align}
\operatorname{log}\mathcal{L}(\theta|x,y) = \sum_{i=1}^N \operatorname{log}\mathcal{L}(\theta|x_i,y_i)
\end{align}

We can take the sum of the log of the output of our `likelihood` method applied to the full dataset to get a better idea of how different $\hat\theta$ compare. We can also plot the different distribution densities over our dataset and see how they line up qualitatively.

```python
theta_hats = [0.5, 1.0, 2.2]
fig, axes = plt.subplots(ncols=3, figsize=(16, 4))
for theta_hat, ax in zip(theta_hats, axes):
  ll = np.sum(np.log(likelihood(theta_hat, x, y))) # log likelihood
  plot_density_image(x, y, theta_hat, ax=ax)
  ax.scatter(x, y)
  ax.set(title=f'theta_hat = {theta_hat}, log likelihood: {ll:.2f}')
```

Using the log likelihood calculation, we see that $\mathcal{L}(\theta=1.0) > \mathcal{L}(\theta=0.5) > \mathcal{L}(\theta=2.2)$.

This is great, now we have a way to compare estimators based on likelihood. But like with the MSE approach, we want an analytic solution to find the best estimator. In this case, we want to find the estimator that maximizes the likelihood.


### Finding the Maximum Likelihood Estimator

Our aim is to find some $\hat\theta$ that maximizes the likelihood function:

\begin{align}
\hat{\theta}_{\textrm{MLE}} = \underset{\theta}{\operatorname{argmax}} \mathcal{L}(\theta|x,y) = \underset{\theta}{\operatorname{argmax}}\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{1}{2\sigma^2}(y-\theta x)^2}.
\end{align}

We know from before that it's much nicer to work with the log likelihood:

\begin{align}
\hat{\theta}_{\textrm{MLE}} = \underset{\theta}{\operatorname{argmax}}\sum_{i=1}^N \operatorname{log}\mathcal{L}(\theta|x_i,y_i).
\end{align}

Substituting in our model and taking its logarithm, we get
\begin{align}
\hat{\theta}_{\textrm{MLE}} = \underset{\theta}{\operatorname{argmax}}-\frac{N}{2} \operatorname{log} 2\pi\sigma^2 - \frac{1}{2\sigma^2}\sum_{i=1}^N (y_i-\theta x_i)^2.
\end{align}

Now, we want to maximize the log likelihood, which is the same as minimizing the negative log likelihood. Just like our solution to minimizing MSE, we can take the derivative of our negative log likelihhood, set it to 0, and solve.

\begin{align}
\frac{\partial\operatorname{log}\mathcal{L}(\theta|x,y)}{\partial\theta}=\frac{1}{\sigma^2}\sum_{i=1}^N(y_i-\theta x_i)x_i = 0
\end{align}

This looks remarkably like the equation we had to solve for the optimal MSE estimator, and, in fact, we arrive to the exact same solution!

\begin{align}
\hat{\theta}_{\textrm{MLE}} = \hat{\theta}_{\textrm{MSE}} = \sum_{i=1}^N \frac{x_i y_i}{x_i^2}
\end{align}

```python
theta_hat_mle = (x @ y) / (x @ x)
fig, ax = plt.subplots()
ll = np.sum(np.log(likelihood(theta_hat_mle, x, y))) # log likelihood
plot_density_image(x, y, theta_hat_mle, ax=ax)
ax.scatter(x, y)
ax.set(title=f'theta_hat = {theta_hat_mle:.2f}, log likelihood: {ll:.2f}')
```

# Summary

- Likelihood vs probability
    - $\mathcal{L}(\theta|x, y) = p(y|\theta, x)$
    - $p(y|\theta, x)$ -> "probability of observing the response $y$ given parameter $\theta$ and input $x$"
    - $\mathcal{L}(\theta|x, y)$ -> "likelihood that parameters $\theta$ produced response $y$ from input $x$"
- Log-likelihood maximization
    - We take the $\textrm{log}$ of the likelihood function for computational convenience
    - The parameters $\theta$ that maximize $\textrm{log}\mathcal{L}(\theta|x, y)$ are the model parameters that maximize the probability of observing the data.
- **Key point**:
  - the log-likelihood is a flexible cost function, and is often used to find model parameters that best fit the data.

# Appendix

### General approach to maximum likelihood estimation

We want to find the parameter value that makes our data set most likely. We have $N$ data points $X = \{x_1,...,x_N\}$ and $Y = \{y_1,...,y_N\}$ and want:

$$\hat{\theta}_{\textrm{ML}} = \underset{\theta}{\operatorname{argmax}} p(Y |X, \theta)$$

We have assumed that the noise affects each output independently, we can therefore factorize as follows:
$$\hat{\theta}_{\textrm{ML}} = \underset{\theta}{\operatorname{argmax}}\prod_{i=1}^N p(y_i |x_i, \theta).$$

Optimizing such a product is numerically unstable (multiplying small values together leads to underflow), so taking the logarithm of the likelihood helps by transforming products into sums:

$$ \textrm{log} \prod_{i=1}^N p(y_i |x_i, \theta) = \sum_{i=1}^m \textrm{log} p(y_i |x_i, \theta)$$

The $\textrm{log}()$ function is *monotonically increasing*, which means that it preserves the order of its inputs. In practice optimiaztion routines are usually developed to solve minimization problem, we therefore rewrite our objective as a minimization of the negative log likelihood. Finally we obtain a problem that is equivalent to the original and more practical to solve:
$$\hat{\theta}_{\textrm{ML}} = \underset{\theta}{\operatorname{argmax}}\prod_{i=1}^m P_{\textrm{model}}(x_i | \theta) = \underset{\theta}{\operatorname{argmin}} -\sum_{i=1}^m \textrm{log}P_{\textrm{model}}(x_i | \theta).$$


---

### Probability vs likelihood

The expression $\mathrm{p}(\mathrm{y} | \mathrm{x}, \theta)$ can be considered as a function of $y$ or of $x$:

1. as a function of $y$, it is the probability distribution of the response conditioned on the input $x$ and the parameter $\theta$ (aka "encoding distribution")
  - Think of **probability** relating to possible **responses**

2. as a function of $\theta$, it is the probability of the data given the model parameter (aka likelihood function, which is not a proabbility distribution, meaning it usually does not sum to one).
  - Think of **likelihood** relating to **parameters**

In order to keep things clean, we can write the encoding distribution as $\mathrm{p}(\mathrm{y} | \mathrm{x}, \theta)$ and the likelihood function as $\mathcal{L}(\theta;x, y)$.



We can also see $\mathrm{p}(\mathrm{y} | \mathrm{x}, \theta)$ as a function of $x$. This is the stimulus likelihood function, and it is useful in case we want to decode the input $x$ from observed responses $y$. This is what is relevant from the point of view of a neuron that does not have access to the outside world and tries to infer what's out there from the responses of other neurons!
